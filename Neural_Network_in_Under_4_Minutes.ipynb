```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
!wget "https://archive.org/download/ages-and-heights/AgesAndHeights.pkl"
```

```python
data = pd.read_pickle('AgesAndHeights.pkl')
data
```

```python
ages = data['Age']
heights = data['Height']
plt.scatter(ages, heights)
plt.title('Height (Inches) vs Age (Years)')
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(8, input_shape=(1,), activation='relu'))
model.add(tf.keras.layers.Dense(1))
model.compile(loss='mse')
```

```python
model.fit(x=ages, y=heights, epochs=1000)
```

```python
plt.scatter(ages, heights)
plt.title('Height (Inches) vs Age (Years)')
x = list(range(18))
preds  = model.predict(x)
plt.plot(x, preds, color='green')
```
