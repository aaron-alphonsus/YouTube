```
import pandas as pd
```

```
pd_df = pd.read_csv('train_and_test2.csv')
pd_df
```

```
useful_cols = pd_df[['Passengerid', 'Age', 'Fare', 'Sex']]
useful_cols
```

```
useful_cols.describe()
```

```
useful_cols.hist()
```

```
age = useful_cols['Age']
age_mean = age.mean()
age_std = age.std()
age_normalized = age.apply(lambda x: (x - age_mean) / age_std)
age_normalized
```

```
useful_cols['AgeNormalized'] = age_normalized
useful_cols
```

```
fare_mean = useful_cols['Fare'].mean()
fare_std = useful_cols['Fare'].std()
useful_cols['FareNormalized'] = useful_cols.apply(lambda row: (row['Fare'] - fare_mean) / fare_std, axis=1)
useful_cols
```

```
useful_cols['AgeNormalized'].hist()
```
